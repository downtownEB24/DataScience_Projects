```python
import matplotlib.pyplot as plt #importing libraries to create plots & charts
import pandas as pd
import os #file directory library

os.chdir("C:/Users/esbro/Desktop/DAAN862/Week 6")#changing file path to point to directory where mtcars is located
mtcars=pd.read_csv('mtcars.csv')
mtcars.head()#outputting first few rows of dataset
```

Question 1.) Plot am-based histogram to compare mpg (20 points)

```python
cars=mtcars.hist('mpg',by='am',bins=6,edgecolor='k',color='coral',density=0)
#creating histogram of cars with different types of am based on mpg
cars[0].set_title("Histogram for Cars with Type AM=0")
cars[1].set_title("Histogram for Cars with Type AM=1")
#creating different titles for the two different AM histograms
for car in cars.flatten():#setting axis labels
    car.set_xlabel("MPG")
    car.set_ylabel("Total number of Cars")
```

Question 2.) Use scatterplot to plot mpg VS. hp (20 points)

```python
plt.style.use('ggplot') #using style of similar aesthetic to ggplot2 package in R
mtcars.plot.scatter(x='hp',y='mpg',s=50)#scatterplot is always y versus x
plt.title("Scatterplot of mpg vs hp")
```

Question 3.) Create a scatterplot matrix for new data consisting of columns [disp, hp, drat, wt, qsec]. (20 points)

```python
#columns [disp, hp, drat, wt, qsec]
new_data=mtcars[['disp','hp','drat','wt','qsec']] #creating new dataframe with only prescribed columns of mtcars
#showing the two styles of scatterplot matrix while showing the kernal density estimation
plt.style.use('classic')
pd.plotting.scatter_matrix(new_data,s=60,diagonal='kde')
plt.suptitle("Classic style Scatterplot matrix") #adding title to the figure
plt.style.use('ggplot')
pd.plotting.scatter_matrix(new_data,s=60,diagonal='kde')
plt.suptitle("GGplot style Scatterplot matrix") #adding title to the figure
```

Question 4.) Create boxplots for new data consisting of columns [disp, hp, drat, wt, qsec]. (20 points)

```python
#columns [disp, hp, drat, wt, qsec]
#dict color is used to map different colors to boxes, whiskers, medians and caps
color=dict(boxes='lightcoral',whiskers='darkred',medians='darkblue',caps='gold')
new_data.plot.box(color=color,sym='r+')
new_data.boxplot()
plt.title("Boxplots for Disp, Hp, Drat, Wt, Qsec")
```
